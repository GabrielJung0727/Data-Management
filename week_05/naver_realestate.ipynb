```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 간단한 5x5 흑백 이미지 행렬 생성 (0~255 값)
image_matrix = np.array([
    [0, 50, 100, 150, 200],
    [50, 100, 150, 200, 255],
    [100, 150, 200, 255, 200],
    [150, 200, 255, 200, 150],
    [200, 255, 200, 150, 100]
])

print("원본 이미지 행렬:")
print(image_matrix)

# 2. 행렬을 벡터로 변환 (평평하게 펴기)
image_vector = image_matrix.flatten()
print("\n행렬을 벡터로 변환:")
print(image_vector)

# 3. 이미지 시각화 (원본)
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(image_matrix, cmap='gray')
plt.title("원본 이미지 (행렬)")
plt.axis('off')

# 4. 밝기 조정 (스칼라 곱): 밝기를 1.5배 증가
brightened_matrix = image_matrix * 1.5
brightened_matrix = np.clip(brightened_matrix, 0, 255)  # 0~255 범위로 클리핑

# 5. 두 이미지 합성 (행렬 덧셈): 원본에 노이즈 추가
noise = np.random.randint(-30, 30, size=image_matrix.shape)
noisy_matrix = image_matrix + noise
noisy_matrix = np.clip(noisy_matrix, 0, 255)  # 0~255 범위로 클리핑

# 6. 결과 시각화
plt.subplot(1, 3, 2)
plt.imshow(brightened_matrix, cmap='gray')
# plt.title("밝기 조정된 이미지")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(noisy_matrix, cmap='gray')
# plt.title("노이즈 추가된 이미지")
plt.axis('off')

plt.tight_layout()
plt.show()

# 7. 행렬 연산 예: 밝기 조정 행렬과 원본 행렬의 곱셈
scalar = 0.8  # 밝기 감소
dimmed_matrix = image_matrix * scalar
dimmed_matrix = np.clip(dimmed_matrix, 0, 255)

print("\n밝기 감소 후 행렬:")
print(dimmed_matrix)

# 8. 벡터 내적 예: 두 벡터 간 유사도 계산 (예: 두 이미지 간의 유사도)
vector1 = image_vector
vector2 = noisy_matrix.flatten()
similarity = np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))  # 코사인 유사도
print("\n두 이미지 벡터 간 코사인 유사도:", similarity)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# 1. 랜덤 2D 데이터 생성 (100개의 샘플, 2차원)
np.random.seed(42)  # 재현성을 위해 시드 설정
data = np.random.randn(100, 2)  # 평균 0, 표준편차 1의 정규 분포 데이터

# 2. PCA를 사용하여 직교 성분 추출 (2차원으로 축소)
pca = PCA(n_components=2)  # 2차원으로 축소
transformed_data = pca.fit_transform(data)

# 3. 주성분 추출
components = pca.components_
print("PCA 주성분 (직교 벡터):")
print(components)

# 4. 주성분 간 직교성 확인 (서로 다른 주성분 벡터 간 내적)
if len(components) >= 2:  # 적어도 2개의 주성분이 있어야 함
    orthogonality_check = np.dot(components[0], components[1])
    print("\n첫 번째와 두 번째 주성분 간 내적 (직교성 확인):", orthogonality_check)

    if np.abs(orthogonality_check) < 1e-10:  # 매우 작은 값이면 직교로 간주
        print("주성분들이 직교합니다 (내적이 0에 가깝습니다).")
    else:
        print("주성분들이 직교하지 않습니다 (내적이 0이 아님).")

# 5. 각 주성분의 노름 확인 (정규화 확인)
norm_component0 = np.linalg.norm(components[0])
norm_component1 = np.linalg.norm(components[1])
print("\n첫 번째 주성분 노름:", norm_component0)  # 약 1이어야 함
print("두 번째 주성분 노름:", norm_component1)  # 약 1이어야 함

# 6. 시각화 (선택 사항, 이전과 동일)
plt.figure(figsize=(8, 4))
plt.scatter(data[:, 0], data[:, 1], alpha=0.5, label="원본 데이터")
for i, component in enumerate(components):
    plt.quiver(0, 0, component[0], component[1], scale=3, color=f'C{i}', label=f'주성분 {i+1}')
plt.title("데이터와 주성분 (직교 벡터)")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.grid(True)
plt.show()
```

```python

# 6. 축소된 데이터 역변환 (원래 공간으로 프로젝션)
reconstructed_data = pca.inverse_transform(transformed_data)

# 7. 결과 시각화
plt.subplot(1, 2, 2)
plt.scatter(reconstructed_data[:, 0], reconstructed_data[:, 1], alpha=0.5, color='red')
plt.title("PCA로 축소 후 복원 (직교 프로젝션)")
plt.xlabel("X1")
plt.ylabel("X2")

plt.tight_layout()
plt.show()

# 8. 설명된 분산 비율 (직교 성분이 데이터 분산을 최대화)
explained_variance_ratio = pca.explained_variance_ratio_
print("\n설명된 분산 비율:", explained_variance_ratio)
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
# nltk 불용어 다운로드 (최초 1회만)
nltk.download('stopwords')
```

```python
from nltk.corpus import gutenberg
nltk.download('gutenberg')

gutenberg.fileids()
```

```python
raw_text = gutenberg.raw('austen-emma.txt')
print(raw_text[:1000])
```

```python
import re
# 1. 영어 단어만 추출 (하이픈 무시)
words = re.findall(r'\b[a-zA-Z]+\b', raw_text)
words[:20]
```

```python
# 2. 소문자로 변환
words = [word.lower() for word in words]
```

```python
'/ '.join(stopwords.words('english'))
```

```python
# 3. 불용어 제거
stop_words = stopwords.words('english')
filtered_words = [word for word in words if word not in stop_words]
'/ '.join(filtered_words)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 4. 단어들을 하나의 문자열로 합치기
filtered_text = ' '.join(filtered_words)

# 5. 워드클라우드 생성
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(filtered_text)

# 6. 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud (without Stopwords)")
plt.show()
```

### gutenberg.fileids() 의 모든 text의 워드클라우드를 그려라
- 함수를 사용하여 반복문으로 순회하며 한 줄에 4개씩 시각화하라

```python
text = gutenberg.fileids()
text
```

```python
stop_words = stopwords.words('english')
for i in ['thee', 'one', 'said', 'must', 'thou', 'went', 'thy', 'much', 'ye', 'unto']:
    stop_words.append(i)

def create_wordcloud(text_file):
    raw_text = gutenberg.raw(text_file)
    words = re.findall(r'\b[a-zA-Z]+\b', raw_text)
    words = [word.lower() for word in words]
    filtered_words = [word for word in words if word not in stop_words]
    filtered_text = ' '.join(filtered_words)
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(filtered_text)
    return wordcloud

text_files = gutenberg.fileids()
num_files = len(text_files)

plt.figure(figsize=(15, 10))  # Adjust figure size for 5 columns

for i, text_file in enumerate(text_files):
    wordcloud = create_wordcloud(text_file)
    plt.subplot(num_files // 4 + (1 if num_files % 4 > 0 else 0), 4, i + 1)  # Create subplots
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(text_file)  # Set title to the filename

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```
